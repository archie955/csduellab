```python
import polars as pl

df = pl.read_csv("./collected_data.csv")

df.head(n=3)
```

```python
def weapon_parse(w):
    if w in {"AK-47", "M4A4", "AUG", "SCAR-20", "M4A1", "FAMAS", "Galil AR", "SG 553"}:
        return "Rifle"
    elif w in {
        "P2000",
        "USP-S",
        "Glock-18",
        "P250",
        "Five-SeveN",
        "Tec-9",
        "CZ75-Auto",
        "CZ75 Auto",
        "Dual Berettas",
        "Desert Eagle",
        "R8 Revolver",
    }:
        return "Pistol"
    elif w in {"Nova", "XM1014", "MAG-7", "Sawed-Off"}:
        return "Shotgun"
    elif w in {"MP9", "MAC-10", "PP-Bizon", "MP7", "UMP-45", "P90", "MP5-SD"}:
        return "SMG"
    elif w in {"SSG 08", "AWP", "SCAR-20", "G3SG1"}:
        return "Sniper"
    elif w in {"M249", "Negev"}:
        return "Machine Gun"
    elif w in {
        "HE Grenade",
        "Molotov",
        "Incendiary Grenade",
        "Smoke Grenade",
        "Decoy Grenade",
        "Flashbang",
        "Flash Grenade",
    }:
        return "Grenade"
    elif w in {"Knife", "Zeus x27", "C4"}:
        return "Melee"
    else:
        return w
```

```python
df = df.with_columns(
    pl.col("attacker_weapon").map_elements(weapon_parse, return_dtype=pl.String)
)
df = df.with_columns(
    pl.col("victim_weapon").map_elements(weapon_parse, return_dtype=pl.String)
)
df.select("attacker_weapon").unique()
df.select("victim_weapon").unique()
```

```python
cols = []
for key in df.collect_schema():
    if key not in {
        "kill_tick",
        "duel_tick",
        "kill_seconds",
        "duel_seconds",
        "attacker_weapon",
        "victim_weapon",
    }:
        cols.append(key)
print(cols)
```

```python
cols = []
for key in df.collect_schema():
    cols.append(key)
print(cols)
```

```python
for col in cols:
    df = df.rename({col: col.replace("attacker", "player_1")})
df.head(n=3)
```

```python
for col in cols:
    df = df.rename({col: col.replace("victim", "player_2")})
df.head(n=3)
```

```python
new_df = df.clone()
new_df.head(n=3)
```

```python
cols = []
for key in new_df.collect_schema():
    cols.append(key)
for col in cols:
    new_df = new_df.rename({col: col.replace("player_1", "temporary")})
new_df.head(n=3)
```

```python
cols = []
for key in new_df.collect_schema():
    cols.append(key)
for col in cols:
    new_df = new_df.rename({col: col.replace("player_2", "player_1")})
new_df.head(n=3)
```

```python
cols = []
for key in new_df.collect_schema():
    cols.append(key)
for col in cols:
    new_df = new_df.rename({col: col.replace("temporary", "player_2")})
new_df.head(n=3)
```

```python
df = df.with_columns(pl.lit(True).alias("player_1_win"))
df.head(n=3)
```

```python
new_df = new_df.with_columns(pl.lit(False).alias("player_1_win"))
new_df.head(n=3)
```

```python
new_df = new_df.with_columns(
    (pl.col("player_2_side") == "ct").alias("player_1_terrorist")
)
new_df = new_df.drop("player_2_side")
new_df.head(n=3)
```

```python
def df_side(s):
    return s == "t"


df = df.with_columns((pl.col("player_1_side") == "t").alias("player_1_terrorist"))
df = df.drop("player_1_side")
df.head(n=3)
```

```python
total_df = pl.concat([df, new_df], how="diagonal")
total_df.head(n=3)
```
